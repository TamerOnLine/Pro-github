## Introduction
In this notebook, we will learn how to use the `MinMaxScaler` tool from the [Scikit-Learn](https://scikit-learn.org/stable/) library to scale data. Data scaling is a crucial step in data preprocessing for machine learning models, ensuring that all features are on the same scale, which can significantly improve the performance of many machine learning algorithms.

### What is Scaling?
-   Scaling is the process of transforming numerical values in a dataset to a specific range, usually between 0 and 1. The goal is to bring all features (variables) onto the same scale so that no single feature dominates the others. For example, in a dataset with columns like age and income, the range of values can be very different (age from 0 to 100 and income from 1,000 to 100,000). If these values are not standardized, the income might overpower the age in some machine learning models, leading to inaccurate results.

-   Simply put, scaling ensures that all columns "play on the same field," meaning all values are within the same range and contribute equally to a machine learning model. This is especially important in models that rely on distances, like K-Nearest Neighbors (KNN) and Support Vector Machines (SVM), where larger values can unfairly influence the results.

**A Simple Example to Illustrate:**

Imagine you are playing a game where you have to compare the heights of different poles, but the heights are given in different units: one pole is 2 meters tall, and another is 2000 centimeters tall. Even though both heights are actually the same, using different measurement units can affect how you perceive them. Scaling here would be like converting all heights to the same unit of measurement and then normalizing them to a range from 0 to 1, making them easier to compare.

# 1. Import Necessary Libraries

[sklearn](https://scikit-learn.org/stable/)

[preprocessing](https://scikit-learn.org/stable/modules/preprocessing.html)

[MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html)

[pandas](https://pandas.pydata.org/)

```python
# Import MinMaxScaler from sklearn.preprocessing for scaling numerical data
# MinMaxScaler is used to transform features to a given range, typically between 0 and 1.
from sklearn.preprocessing import MinMaxScaler

# Import pandas library for data manipulation and analysis
# pandas provides powerful data structures like DataFrames to work with structured data efficiently.
import pandas as pd
import os 
```

-   pandas: A popular library in Python for data analysis and manipulation.
-   MinMaxScaler: A tool from the scikit-learn library used for scaling numerical data to a specified range, usually between 0 and 1. [up](#data-scaling-with-minmaxscaler-using-scikit-learn)

# 2. Load Data from a CSV File

[read_csv](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

```python
# Read a CSV file named 'bank.csv' into a pandas DataFrame
# Make sure the file is present in the same directory as the notebook
df = pd.read_csv('/kaggle/input/minmax-data/bank.csv')
```

-   The read_csv function from the pandas library is used to load data from a CSV file named 'bank.csv' into a DataFrame object called df.[up](#data-scaling-with-minmaxscaler-using-scikit-learn)

# 3. Selecting Numerical Columns from the DataFrame

[select_dtypes](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.select_dtypes.html)

[columns](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.columns.html)

```python
# Select columns with numerical data types from the DataFrame
# Using the select_dtypes() method to filter columns with only numeric data types
# 'include='number'' specifies that only columns with numeric data types (e.g., int, float) should be selected
numerical_columns = df.select_dtypes(include='number').columns
```

-   This line selects only the numerical columns from the **DataFrame** named **df**.
-   The **select_dtypes()** function is used to filter columns by their data type.
-   **include=[number]** specifies that we only want columns with numerical data types (like integers and floats).
-   **.columns** retrieves the names of these columns.

# 4. Extracting the Numerical Data Using the Identified Numerical Columns

```python
# Create a new DataFrame containing only the numerical columns from the original DataFrame
# Using the list of numerical column names (numerical_columns) to filter the original DataFrame
# This helps in focusing on numerical data for tasks such as scaling, normalization, or machine learning
data = df[numerical_columns]
```

-   This line creates a new **DataFrame** named **data** containing only the numerical columns from the original **df**.
-   The original DataFrame (**df**) is indexed with **numerical_columns** to select only the desired columns.

# 5. Creating a MinMaxScaler Object

```python
# Initialize a MinMaxScaler object from sklearn.preprocessing
# MinMaxScaler scales and translates each feature individually such that it is in the given range (usually between 0 and 1)
# This is useful for normalizing numerical data to improve the performance of machine learning algorithms
scaler = MinMaxScaler()
```

-   This line initializes a **MinMaxScaler** object from the **sklearn.preprocessing** module.
-   The MinMaxScaler scales and translates each feature in the DataFrame such that it is in the range [0, 1].
-   This is often done to normalize data before feeding it into machine learning algorithms.

# 6. Fit and Transform the Data Using MinMaxScaler

[scaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)

```python
# Apply fit_transform() to scale the data using MinMaxScaler
# fit() computes the minimum and maximum values for each feature in the data
# transform() then scales the data using these computed min and max values
# The result is a NumPy array containing the scaled data
scaled_data = scaler.fit_transform(data)
```

-   The **fit_transform()** method is called on the scaler object.
-   **fit()** computes the minimum and maximum values for each column in data.
-   **transform()** then scales the data using these minimum and maximum values.
-   The result, **scaled_data**, is a NumPy array where each value is scaled to fall between 0 and 1.

# 7. Creating a New DataFrame from the Scaled Data

```python
# Create a new DataFrame from the scaled data using the original numerical column names
# The scaled_data, which is a NumPy array resulting from the MinMaxScaler transformation, is converted back to a pandas DataFrame
# The 'columns' parameter is used to set the original numerical column names for the new DataFrame to maintain consistency and readability
scaled_df = pd.DataFrame(scaled_data, columns=numerical_columns)
```

-   This line creates a new **DataFrame** named **scaled_df** from the **scaled_data** array.
-   **columns=numerical_columns** ensures the new DataFrame retains the same column names as the original numerical columns from **df**.
-   This keeps the DataFrame clear and easier to interpret by retaining the original column labels.[up](#data-scaling-with-minmaxscaler-using-scikit-learn)

# 8. Display Data

- Displaying Data Before Scaling:


```python
# Display the original numerical data before MinMax Scaling
# This shows the data in its original scale before any transformation is applied
print('Data before MinMax Scaling:')
data.round(4)  # Display the original data rounded to 4 decimal places for better readability

```

- Displaying Data After Scaling:

```python
# Print a message to indicate that the data shown is after MinMax Scaling
print('\nData after MinMax Scaling:')

# Display the scaled data rounded to 4 decimal places
# The round() method is used to round the values in the DataFrame to the specified number of decimal places.
# This improves readability and makes it easier to understand the scaled values.
scaled_df.round(4)
```

# **Codes for Displaying Outputs and Visualizations**

```python
# Import matplotlib for plotting and seaborn for advanced visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Displaying Data After Scaling with MinMaxScaler:
- You can display some statistics of the data after scaling using the describe() function.

```python
# Plot the distribution of the original data before scaling
# plt.figure() sets the size of the plot. Here, we set the figure size to 12x6 inches.
# sns.histplot() is used to plot a histogram of the original numerical data. 'kde=True' adds a Kernel Density Estimate line.
# 'element='step'' changes the style of the histogram bars for better visualization.
plt.figure(figsize=(12, 6))
sns.histplot(data, kde=True, element='step')
plt.title('Distribution of Original Data')  # Title of the plot
plt.show()  # Display the plot
```

2. Visualizing the Distribution Before and After Scaling:
- Use the seaborn library to create plots to compare the distribution before and after scaling.

```python
# Plot the distribution of the data after scaling
# Another histogram is plotted using the scaled data to compare its distribution with the original data.
plt.figure(figsize=(12, 6))
sns.histplot(scaled_df, kde=True, element='step')
plt.title('Distribution of Scaled Data')  # Title of the plot
plt.show()  # Display the plot
```

3. Comparing Original and Scaled Values Using Boxplots:

```python
# Boxplot of the original data
# Boxplots are used to show the distribution, spread, and outliers of numerical data.
# plt.figure() sets the size of the plot to 12x6 inches.
# sns.boxplot() creates a boxplot for each numerical column in the DataFrame 'data'.
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.title('Original Data Boxplot')  # Set the title of the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()  # Display the plot
```

```python
# Boxplot of the scaled data
# Another boxplot is created for the scaled data to visualize how scaling affects the data distribution.
plt.figure(figsize=(12, 6))
sns.boxplot(data=scaled_df)
plt.title('Scaled Data Boxplot')  # Set the title of the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()  # Display the plot
```

4. Displaying Relationships Between Variables Using Pairplot:

- You can use seaborn to plot a pairplot to visualize relationships between variables.

```python
# Pairplot of the scaled data
# A pairplot is a grid of scatter plots for each pair of numerical features in the DataFrame.
# It helps in visualizing the relationships between all numerical variables and identifying patterns, trends, and potential correlations.
# By default, the diagonal plots are histograms that show the distribution of individual variables.
sns.pairplot(scaled_df)

# Display the pairplot
plt.show()
```

1. Pairplot of the Scaled Data:

-   The code uses the sns.pairplot() function from the Seaborn library to create a pairplot of the scaled_df DataFrame, which contains the scaled numerical data.
-   A pairplot is a grid of scatter plots, where each scatter plot represents the relationship between a pair of numerical features (columns) in the DataFrame.
-   Pairplots are useful for visualizing relationships between all numerical variables in a dataset, allowing you to see patterns, trends, potential correlations, and even outliers.
-   By default, the diagonal of the pairplot contains histograms that display the distribution of individual variables, helping to understand their spread and skewness.

2. Displaying the Pairplot:

-   plt.show() is used to display the pairplot. This function is from the Matplotlib library, and it renders the plot in the output area.
-   The visualization helps in exploratory data analysis (EDA) by providing insights into how different numerical features relate to each other in the scaled dataset.
